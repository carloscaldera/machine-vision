```python
image_data = data.camera() # load test image

print('Test Image shape : {}'.format(image_data.shape))
print('Test Image dtype : {}'.format(image_data.dtype)) # ranges from 0-255 

plt.figure(dpi=200)
plt.imshow(image_data, cmap=plt.cm.gray) # before
plt.show()
```

# Inverting an image

$$ INVERT : [[ Q0 = 255 - P0; ]]$$

```python
plt.figure(dpi=200)
plt.imshow(255 - image_data, cmap=plt.cm.gray) # after
plt.show() 
```

# Stretching the contrast of an image

$$ STRETCH : [[ Q0 = P0 * gamma + beta; ]] $$

```python
stretched_image = np.clip(image_data*1.4 + 1, 0, 255)
plt.figure(dpi=200)
plt.imshow(stretched_image, cmap=plt.cm.gray) # where gamma is greater than one 
plt.show()
```

# Thresholding

$$ THRESH: [[ if (P0 > thresh) A0 =1; else A0 = 0; ]]  $$ 

```python
threshold_image = np.zeros(image_data.shape, dtype=np.bool_) # create an array of the same shape as test image
threshold = 50 # arbitrary threshold
threshold_image[ image_data >= threshold] = 1 # threshold image, zeros are already taken into account 

plt.figure(dpi=200)
plt.imshow(threshold_image, cmap=plt.cm.gray)
plt.show()
```
